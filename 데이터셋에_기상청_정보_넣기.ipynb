```python
from geopy.distance import geodesic
```

```python
def haversine_distance(lat1, lon1, lat2, lon2):
  return geodesic((lat1,lon1),(lat2,lon2)).miles
```

```python
address = [[37.57142,126.9658,'서울'],[37.47772,126.6249,'인천'],[37.90188,127.0607,'동두천'],[37.88589,126.76648,'파주'],[37.25746,126.983,'수원'],[37.48863,127.49446,'양평'],[37.26399,127.48421,'이천']]
```

```python
for i in range(len(df)):
  loc = 9999
  place = ''
  for j in range(len(address)):
    temp = haversine_distance(df.iloc[i]['GRID_Y'], df.iloc[i]['GRID_X'], address[j][0], address[j][1])
    if temp < loc:
      loc = temp
      place = address[j][2]
  df.loc[i,'address'] = place
```
